```python
# # Broodcasting
# [100,200,300] in taka  ---> flat 10% discount
prices= [100,200,300]
discount=10
final_prices=[]
for price in prices:
    final_price=price-(price*discount/100)
    final_prices.append(final_price)
print(final_prices)
# for time complexity problem in loop we are using broadcasting
```

```python
# Handling Missing values
#why ---->
#buildin funtion
# np.isnan() ---> detect missing values
# np.nan_to_num()
#np.isinf() ----> detecting infinite values
#nan --> not a number  0/0  it returns boolean value
#np.isnun(array)
import numpy as np
arr=np.array([1,2,np.nan,4,np.nan,6])
print(np.isnan(arr))
print(np.nan==np.nan) # nehi we can not compare them directly
```

```python
# nan_to_num
# np.nan_to_num(array,nan=value) defalt -0
import numpy as np
arr=np.array([1,2,np.nan,4,np.nan,6])
cleaned_arr=np.nan_to_num(arr,nan=4)
print(cleaned_arr)
```

```python
#handling infinite value
#np.isinf(array) 10^1000
#1/0
import numpy as np
arr=np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
```

```python
#replace
import numpy as np
arr=np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
cleaned_arr=np.nan_to_num(arr,posinf=1000,neginf=-1000)
print(cleaned_arr)
```
