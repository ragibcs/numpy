```python
#filtering rows salaries greate rthan 50 and age is greater than 40
import pandas as pd
data={
    "Name":['Ragib','hasan','leetcode',"Rian","luna","muna","Mayan","July","Ruchi","Shuvo"],
    "Age": [20,20,21,23,46,76,45,56,44,32],
    "Salary":[50000,60000,70000,80000,80000,90000,10000,150000,70000,80000],
    "Performance_Score":[85,90,56,98,89,80,77,79,67,80]
}
df=pd.DataFrame(data)
con1=df["Salary"]
con2=df["Age"]
con3=df[(con1>60000) & (con2>44)]
print(con3)


print("\n After a or condition")
# salary greater thn 70000 or performance_Score > 77
print(df[(df["Salary"]>70000) | (df["Performance_Score"]>60)])
```
